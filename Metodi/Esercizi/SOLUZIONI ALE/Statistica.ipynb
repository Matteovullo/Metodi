```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
X = np.loadtxt("data_altezze.dat")

n= 2000
print("media..: " , X.mean())
print("deviazione standard :" , np.std(X)) #utilizzare ddof=1 se avevamo solo una parte della campione
```

```python
plt.hist(X,20, edgecolor="black")
plt.show()
```

```python
#è possibile adattare ai dati una distribuzione normale?
from scipy.stats import probplot,norm
fig, ax = plt.subplots(1,1)
probplot(X,dist=norm,plot=ax)
plt.show()
```

```python
#per verificare che i dati utilizzano la distribuzione normale abbiamo:

mu3= np.sum((X - np.mean(X))**3)/X.size
mu2= np.sum((X - np.mean(X))**2)/X.size
mu4= np.sum((X - np.mean(X))**4)/X.size


b1 = mu3**2/mu2**3
b2= mu4/mu2**2

print("b1: ",b1," b2: ",b2)
print("dato che b1 è vicino a 0 e b=2 tende a 3 allora i dati seguono una distribuzione normale")
```

```python
from scipy.stats import chi2
X = np.array([2.05, 2.04, 1.98, 1.96, 2.03, 2.01, 1.97, 1.99, 2.01, 2.05, 1.96, 1.95, 2.04, 2.01, 1.97, 1.96, 2.02, 2.04, 1.98, 1.94])
N = X.size
S = np.std(X, ddof=1)
alfa= 0.05

chi1= chi2.ppf(1-alfa/2, N-1)
chi2= chi2.ppf(alfa/2, N-1)

I_l= (S**2*(N-1))/(chi1)
I_r= (S**2*(N-1))/(chi2)

print("i_r e I_l : ", I_r , I_l)
```
